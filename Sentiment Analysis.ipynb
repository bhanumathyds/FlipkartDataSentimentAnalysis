```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('brown')
nltk.download("wordnet")
from textblob import TextBlob
```

```python
with open('text.txt', 'r') as f:
    text = f.read()
```

```python
my_sentence = TextBlob(text)

pos_tags = my_sentence.tags
```

```python
# Noun phrase extraction
noun_phrases = my_sentence.noun_phrases
print("Noun Phrases:", noun_phrases)

# Sentiment analysis
sentiment = my_sentence.sentiment
print("Sentiment Polarity:", sentiment.polarity)
print("Sentiment Subjectivity:", sentiment.subjectivity)

# Tokenization
tokens = my_sentence.words
print("Tokens:", tokens)

# Word inflection and lemmatization
inflected_words = [word.pluralize() for word in tokens]
lemmatized_words = [word.lemmatize() for word in tokens]
```

```python
# Create a TextBlob object
blob  = TextBlob(text)

# Extract nouns
nouns = [word for word, pos in blob.tags if pos == 'NN' or pos == 'NNS'] # NN singular noun, NNS plural noun
print(nouns)

# Count the frequency of each noun
noun_frequency = {}
for noun in nouns:
    noun_frequency[noun] = noun_frequency.get(noun, 0) + 1

# print noun frequency
for noun, frequency in noun_frequency.items():
    print(f"Noun: {noun}, Frequency: {frequency}")
```

```python
# Create a TextBlob object
blob = TextBlob(text)

# Perform sentiment analysis
sentiment = blob.sentiment

print("Sentiment Polarity: ", sentiment.polarity)
print("Sentiment Subjectivity: ", sentiment.subjectivity)

# Interpret the sentiment polarity
# higher value more subjective
# lower value more objective - 1 to 0
# 0 will indicate neutral sentiment

if(sentiment.polarity > 0):
    print("The text is Positive.")
elif(sentiment.polarity < 0):
    print("The text is negative.")
else:
    print("The text is Neutral.")
```
